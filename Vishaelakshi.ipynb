```python
import pandas as pd
df = pd.read_csv('Train.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df['Outlet_Type'].unique()
```

```python
df['Item_Fat_Content'].unique()
```

```python
df['Item_Fat_Content'].replace({'low fat': 'LF','Low Fat': 'LF','Regular':'Reg','reg':'Reg'}, inplace=True)
```

```python
df['Item_Fat_Content'].unique()
```

```python
df.drop(columns =['Item_Weight','Item_Identifier','Item_Visibility','Outlet_Identifier','Outlet_Establishment_Year'],inplace=True)
```

```python
df['Outlet_Size'].fillna('Small',inplace=True)
```

```python
df.head()
```

```python
df['Item_Fat_Content'].replace({'LF':0,'Reg':1},inplace=True)
```

```python
df.head()
```

```python
enc_Outlet_lt = pd.get_dummies(df['Outlet_Location_Type'])
enc_Outlet_lt.drop(columns='Tier 3', inplace=True)
enc_Outlet_lt.head()
```

```python
enc_Outlet_sz = pd.get_dummies(df['Outlet_Size'])
enc_Outlet_sz.drop(columns='Small',inplace=True)
enc_Outlet_sz.head()
```

```python
enc_Outlet_tp = pd.get_dummies(df['Outlet_Type'])
enc_Outlet_tp.drop(columns='Grocery Store',inplace=True)
enc_Outlet_tp.head()
```

```python
df.drop(columns=['Outlet_Type','Outlet_Size','Outlet_Location_Type'],inplace=True)
df = pd.concat([df,enc_Outlet_lt,enc_Outlet_tp,enc_Outlet_sz],axis=1)
df.head()
```

```python
df.drop(columns='Item_Type',inplace=True)
```

```python
X = df.drop(columns='Item_Outlet_Sales')
y = df['Item_Outlet_Sales']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
clf = RandomForestRegressor()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
import xgboost as xg
xgb_r = xg.XGBRegressor(objective ='reg:linear',n_estimators = 10, seed = 123)
```

```python
xgb_r.fit(X_train, y_train)

y_pred = xgb_r.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```
